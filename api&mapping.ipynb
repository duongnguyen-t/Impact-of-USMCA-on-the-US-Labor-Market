```python
# Perform the merge
trade_conc = dftrade_19.merge(
    df_concordance[['hs8', 'hs6', 'naics', 'commodity']],
    left_index=True,
    right_on='commodity',
    how='right',
    indicator=True
)

print(trade_conc.head())
```

```python
#groupby and assign naics
def concordance_group(df):
    
    #print(df.index)
    
    num_naics = df.shape[0]
    # This tells me how many naics codes are there
    
    if num_naics == 1:
        # if only one (like most), just grab the naics code
        
        max_naics = df.naics.iloc[0]
        
    else:
        # if many, then grab the max trade flows...
        
        max_naics = df.loc[df.total_trade.idxmax()].naics
    
    foo = {"num_naics": [num_naics],
          "naics": [max_naics]}
    
    return pd.DataFrame(foo)
```

### Tariff 

```python
tariff = pd.read_csv("tariff.csv")
```

```python
tariff.head()
```

```python
tariff.rename(columns={'hts8': 'hs8'}, inplace=True)
```

```python
# Convert hs8 to string and add leading zero to hs8 values
tariff['hs8'] = tariff['hs8'].astype(str).str.zfill(8)
tariff.head()
```

```python
trade_tariff_20 = mxtrade20.merge(tariff, on='hs8', how='left')

trade_tariff_20.head()
```

```python
# Drop hs6_y, rename hs6_x to hs6, drop _merge, drop E_COMMODITY
trade_tariff_20.drop(columns=['hs6_y', '_merge'], inplace=True)
trade_tariff_20.rename(columns={'hs6_x': 'hs6'}, inplace=True)

# Create naics3 from naics for 0:3 and name it industry code
trade_tariff_20['industry_code'] = trade_tariff_20['naics'].str[0:3]

trade_tariff_20.head(15)
```

```python
rate_columns = ['mfn_ad_val_rate', 'mfn_specific_rate', 'mexico_ad_val_rate', 'mexico_specific_rate']
trade_tariff_20[rate_columns] = trade_tariff_20[rate_columns].fillna(0)
trade_tariff_20.head()
```

```python
# Calculate the percentage change in trade from 2019 to 2020
trade_tariff_20['change'] = trade_tariff_20['mexico_trade'] - mxtrade19['mexico_trade']
trade_tariff_20['% change'] = ((trade_tariff_20['mexico_trade'] - mxtrade19['mexico_trade']) / mxtrade19['mexico_trade']) * 100

# Rearrange the order of columns
columns_order = ['commodity', 'hs8', 'hs6', 'naics', 'industry_code', 'total_trade', 'mexico_trade', 'change','% change', 
                 'mfn_ad_val_rate', 'mfn_specific_rate', 'mexico_ad_val_rate', 'mexico_specific_rate', 'begin_effect_date']

trade_tariff_20 = trade_tariff_20[columns_order]

trade_tariff_20.head()
```

```python
# Calculate Mexico's share of total trade in good j
trade_tariff_20['mexico_share_good_j'] = trade_tariff_20['mexico_trade'] / trade_tariff_20['total_trade']

# Calculate Mexico's share in total trade
total_mexico_trade = trade_tariff_20['mexico_trade'].sum()
total_trade = trade_tariff_20['total_trade'].sum()
mexico_share_total_trade = total_mexico_trade / total_trade

# Calculate RCA
trade_tariff_20['RCA'] = trade_tariff_20['mexico_share_good_j'] / mexico_share_total_trade

trade_tariff_20.head()
```

```python
trade_tariff_20['RCA_compare'] = (trade_tariff_20['RCA'] > 1).astype(int)

trade_tariff_20.head()
```
